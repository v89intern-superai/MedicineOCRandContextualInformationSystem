```python
import os
import cv2
import numpy as np
from PIL import Image
import pytesseract
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
# --- Preprocessing ---
img = cv2.imread("../images/para.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.fastNlMeansDenoising(gray, h=12)

thresh = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY, 15, 11
)
clean = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, np.ones((1, 1), np.uint8))

# Convert to PIL image for pytesseract
final = Image.fromarray(clean)
final
```

```python
# --- OCR with Tesseract ---
# 'tha' = Thai, 'eng' = English
text = pytesseract.image_to_string(final, lang='tha+eng')

print(text)
```
